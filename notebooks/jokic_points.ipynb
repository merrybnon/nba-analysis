# Nikola Jokic — points per game distribution

Data is fetched from NBA.com via `nba_api` and cached locally under `data/cache/`,
so re-running is fast and doesn't hammer the stats server.

Season convention: the ending year — `2025` = the 2024-25 season.

```python
import matplotlib.pyplot as plt
import seaborn as sns

from nba_analysis import get_game_logs, player_stat_dist

sns.set_theme(style="whitegrid")
```

## Single season

```python
player_stat_dist("Nikola Jokic", 2025, binwidth=3)
plt.show()
```

## Multiple seasons pooled together

```python
player_stat_dist("Nikola Jokic", [2022, 2023, 2024, 2025], binwidth=3)
plt.show()
```

## Same seasons, colored by season to see the year-over-year shift

```python
player_stat_dist("Nikola Jokic", [2023, 2024, 2025], binwidth=3, hue_by_season=True, kde=False)
plt.show()
```

## Generalizes to any player and any game-log stat

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
player_stat_dist("Shai Gilgeous-Alexander", 2025, binwidth=3, ax=axes[0])
player_stat_dist("Nikola Jokic", 2025, stat="AST", binwidth=2, ax=axes[1])
plt.tight_layout()
plt.show()
```

## The raw game log is available too, for anything beyond plotting

```python
df = get_game_logs("Nikola Jokic", 2025)
df[["GAME_DATE", "MATCHUP", "WL", "MIN", "PTS", "REB", "AST"]].tail(10)
```
